```python
import matplotlib.pyplot as plt
import seaborn as sns

def plot_bar_chart(df, column):
    counts = df[column].value_counts()
    plt.figure(figsize=(10, 6))
    sns.barplot(x=counts.index, y=counts.values, palette='viridis')
    plt.title(f'Distribution of {column}')
    plt.xlabel(column)
    plt.ylabel('Count')
    plt.show()

# Example: Plot distribution of 'Phase'
plot_bar_chart(df, 'Phase')
```

```python
import warnings
warnings.filterwarnings("ignore", category=UserWarning, module="matplotlib")

def scatterplot(df, x_column, y_column, hue_column):
    """
    Function to plot a scatter plot with specified x, y, and hue columns.

    Returns a scatter plot.
    """
    plt.figure(figsize=(8,6))
    sns.scatterplot(data=df, x=x_column, y=y_column, hue=hue_column)
    plt.title(f'Scatter Plot of {x_column} vs {y_column} with Hue based on {hue_column}')
    plt.xlabel(x_column)
    plt.ylabel(y_column)
    plt.legend(title=hue_column)
    plt.show()

# Example usage with your dataset and specific columns
# Replace 'df' with your actual DataFrame and adjust column names as per your dataset
scatterplot(df, x_column='Constituency_No', y_column='Age', hue_column='Phase')
```

```python
def plot_correlation_heatmap(df):
    plt.figure(figsize=(6, 4))
    sns.heatmap(df.corr(), annot=True, cmap='coolwarm', linewidths=0.5)
    plt.title('Correlation Matrix Heatmap')
    plt.show()
# Show the heatmap
plot_correlation_heatmap(numeric_df)

```
